```python
import json
import geopandas as gpd
import os
from planet import Auth, Session
```

```python
#path_to_orders='/explore/nobackup/people/almullen/smallsat_augmentation/data/orders/YKD/Ch010v026' #quinhagak
path_to_orders='/explore/nobackup/people/almullen/smallsat_augmentation/data/orders/YKD/Ch009v024' #kwigillingok
path_to_api_key = '/explore/nobackup/people/almullen/smallsat_augmentation/code/PL_API_KEY.txt'

#Authenticate Planet
with open(path_to_api_key) as f:
    pl_api_key = f.read(1)
    auth = Auth.from_key(pl_api_key)
    auth.store()
```

```python
#place the request with asynchronous function
async def download(order_id, download_dir):
    async with Session() as sess:
        cl = sess.client('orders')
        await cl.download_order(order_id, directory=download_dir)
```

```python
for order in os.listdir(path_to_orders):
    
    order_id = os.path.join(path_to_orders, order, 'order_id.txt')
    download_dir = os.path.join(path_to_orders, order)
    
    with open(order_id, 'r') as file:
        
        order_id = file.read()
        if not os.path.isdir(os.path.join(download_dir, order_id)):
            print(f'downloading: {order_id} to {download_dir}')
            await download(order_id, download_dir)
        else:
            print(f'{order_id} already downloaded, skipping')
        
```
